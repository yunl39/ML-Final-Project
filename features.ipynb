```python
import pandas as pd
```

```python
all_posts = pd.read_csv("posts.csv")
```

```python
text = all_posts[["Created On", "Post Text"]]
text.columns = ["Date", "Text"]
text.Date = pd.to_datetime(text.Date)
text['Date'] = text['Date']
```

```python
text["Tariff"] = text.Text.str.contains(
    r"\b(tariff|tariffs|trade war|import tax|export tax|customs duties|trade barrier|trade sanctions|import restrictions|tariff hike|retaliatory tariffs|tariff exemption)\b", 
    case=False, regex=True) * 1
```

```python
text["AAPL"] = text.Text.str.contains(
    r"\b(apple|aapl|iphone sales|mac sales|apple earnings|apple stock|apple revenue|apple profit|tim cook|apple forecast|iphone demand|apple supply chain|apple lawsuit|apple buyback|apple dividend)\b", 
    case=False, regex=True) * 1

text["TSLA"] = text.Text.str.contains(
    r"\b(tesla|tsla|elon musk|tesla earnings|tesla stock|tesla deliveries|tesla production|model 3 sales|cybertruck|tesla gigafactory|tesla lawsuit|tesla price cut|autopilot investigation)\b", 
    case=False, regex=True) * 1

text["META"] = text.Text.str.contains(
    r"\b(meta|facebook|meta earnings|fb stock|instagram revenue|meta layoffs|metaverse investment|oculus sales|meta advertising slowdown|meta regulation|whatsapp growth)\b", 
    case=False, regex=True) * 1

text["AMZN"] = text.Text.str.contains(
    r"\b(amazon|amzn|aws growth|amazon earnings|amazon prime memberships|amazon stock|amazon layoffs|amazon labor strike|amazon revenue forecast|amazon advertising)\b", 
    case=False, regex=True) * 1

text["NVDA"] = text.Text.str.contains(
    r"\b(nvidia|nvda|nvidia earnings|nvidia stock|nvidia ai chips|gpu sales|h100 demand|nvidia supply chain|nvidia lawsuit|semiconductor boom|ai chip market)\b", 
    case=False, regex=True) * 1

text["Tech"] = text.Text.str.contains(
    r"\b(tech sector|technology stocks|semiconductor shortage|ai boom|it sector growth|tech layoffs|software spending|hardware demand|cloud growth|cybersecurity threat)\b", 
    case=False, regex=True) * 1

text["Consumer"] = text.Text.str.contains(
    r"\b(consumer spending|retail sales|consumer sentiment|e-commerce boom|holiday shopping|inflation impact on consumers|consumer goods demand|consumer discretionary sector|consumer)\b", 
    case=False, regex=True) * 1

text["Com"] = text.Text.str.contains(
    r"\b(communication services sector|telecom growth|social media ad revenue|messaging app expansion|internet services growth|wireless carrier competition|communication)\b", 
    case=False, regex=True) * 1

text["Tariff"] = text.Text.str.contains(
    r"\b(tariff|tariffs|trade war|import tax|export tax|customs duties|trade barrier|trade sanctions|import restrictions|tariff hike|retaliatory tariffs|tariff exemption)\b", 
    case=False, regex=True) * 1
```

```python
text.dropna(inplace = True)
```

```python
for col in ["AAPL", "TSLA", "META", "AMZN", "NVDA", "Tech", "Consumer", "Com"]:
    print(text[col].sum())
```

```python
import re
text['Text'] = text['Text'].apply(lambda x: re.sub(r'http\S+|www\S+', '', x))
text['Text'] = text['Text'].str.replace(r'[^A-Za-z\s]', '', regex=True)
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words='english',
    token_pattern=r'\b[a-zA-Z]{2,}\b'
)

tfidf_matrix = vectorizer.fit_transform(text['Text'].fillna(''))
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
tfidf_df.index = text.Date
```

```python
tfidf_df = tfidf_df.reset_index()
```

```python
tfidf_df.Date = tfidf_df.Date.dt.date
```

```python
day_sum  = tfidf_df.groupby("Date").sum()
```

```python
text_day = text.drop(columns="Text")
text_day.Date = text_day.Date.dt.date
text_day = text_day.groupby("Date").sum()
```

```python
text_day['tfidf_sum'] = day_sum.sum(axis=1)
text_day['tfidf_mean'] = day_sum.mean(axis=1)
text_day['tfidf_max'] = day_sum.max(axis=1)
important_words = ['growth', 'inflation', 'profit', 'revenue', 'rate', 'debt', 'earnings', 'guidance', 'market', 'forecast', 'downgrade', 'upgrade']
existing_words = [word for word in important_words if word in day_sum.columns]
text_day['important_words_sum'] = day_sum[existing_words].sum(axis=1)
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
text['sentiment'] = text['Text'].fillna('').apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
selected = all_posts[["Created On", "Score", "Upvote Ratio", "Total Comments"]]
selected["sentiment"] = text.sentiment
```

```python
selected.rename({"Created On" : "Date"}, axis = 1, inplace= True)
selected = selected.dropna()
selected.Date = pd.to_datetime(selected.Date).dt.date
```

```python
# Step 1: Compute weighted sentiment
selected["weighted_sentiment"] = selected["sentiment"] * selected["Score"]

# Step 2: Compute weighted average sentiment per day
weighted_sentiment = selected.groupby("Date").apply(
    lambda x: x["weighted_sentiment"].sum() / x["Score"].sum() if x["Score"].sum() > 0 else 0
).reset_index(name="Weighted Sentiment")

# Step 3: Basic daily aggregation
daily_summary = selected.groupby("Date").agg({
    "Score": "sum",
    "Upvote Ratio": "mean",
    "Total Comments": "sum"
}).reset_index()

# Step 4: Add post count
daily_summary["Post Count"] = selected.groupby("Date").size().values

# Step 5: Merge weighted sentiment into summary
daily_summary = daily_summary.merge(weighted_sentiment, on="Date", how="left")
```
